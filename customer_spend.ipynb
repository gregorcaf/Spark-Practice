```python
from pyspark import SparkConf, SparkContext
```

```python
# configures the SparkContext => local machine and not cluster, sets the app name
conf = SparkConf().setMaster("local").setAppName("CustomerSpend")
sc = SparkContext(conf = conf)
sc
```

```python
def parser(line):
    line = line.split(",")
    customer_id = int(line[0])
    price = float(line[2])
    return (customer_id, price)
```

```python
rdd = sc.textFile("customer-orders.csv").map(parser)
```

```python
results = rdd.reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], ascending = False)
results.collect()
```
